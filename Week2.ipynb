```python
import pandas as pd
import glob
```

```python
# Creating dataframe with Listing Files
listing_files = glob.glob("Listing/CRMLSListing*.csv")
listed_df = pd.concat([pd.read_csv(f)for f in listing_files], ignore_index=True)

# Creating dataframe with Sold Files
sold_files = glob.glob("Sold/CRMLSSold*.csv")
sold_df = pd.concat([pd.read_csv(s)for s in sold_files], ignore_index=True)
```

```python
# Row and Column Counts
# Listing Df Shape
print("Listing Df Shape: ", listed_df.shape)

# Sold Df Shape
print("Sold Df Shape: ", sold_df.shape)

# Columns
# Listing Columns
print("Listing Columns: ", listed_df.columns)

# Sold Columns
print("Sold Columns: ", sold_df.columns)

# Datatypes
# Listing Datatypes
print("Listing Datatypes: ", listed_df.dtypes)

# Sold Datatypes
print("Sold Datatypes: ", sold_df.dtypes)
```

```python
# Checking Property Categories

# Unique Property Type Listing Df
print("Unique Property Type Listing Df: ", listed_df["PropertyType"].unique())

# Unique Property Type Sold Df
print("\nUnique Property Type Sold Df: ", sold_df["PropertyType"].unique())
```

```python
# Filter Residential

# Listed df
listed_df = listed_df[listed_df["PropertyType"] == "Residential"]

#Sold df
sold_df = sold_df[sold_df["PropertyType"] == "Residential"]
```

```python
#Check Null Count and Percentages

null_listed = (listed_df.isnull().sum() / len(listed_df)) * 100
null_sold = (sold_df.isnull().sum() / len(sold_df)) * 100

print("Listed Null Percentage:\n", null_listed.sort_values(ascending=False).head(10))
print("\nListed Sold Percentage:\n", null_sold.sort_values(ascending=False).head(10))
```

```python
# Drop Null Values

# For listed
non_null_listed = int(len(listed_df) * 0.10)
listed_df = listed_df.dropna(axis=1, thresh=non_null_listed)

# For sold
non_null_sold = int(len(sold_df) * 0.10)
sold_df = sold_df.dropna(axis=1, thresh=non_null_sold)
```

```python
#Residential vs. other property type share

property_share = sold_df['PropertyType'].value_counts(normalize=True) * 100
print(property_share)
```

```python
# Median and average close price
print("Median Close Price:", sold_df['ClosePrice'].median())
print("Average Close Price:", sold_df['ClosePrice'].mean())
```

```python
# Days on Market distribution
print(sold_df['DaysOnMarket'].describe())
```

```python
# % sold above vs. below list price
sold_df['above_list'] = sold_df['ClosePrice'] >= sold_df['ListPrice']
print(sold_df['above_list'].value_counts(normalize=True) * 100)
```

```python
#Counties with highest median prices
county_median = sold_df.groupby('CountyOrParish')['ClosePrice'].median()
print(county_median.sort_values(ascending=False).head(10))
```
